```python
import pandas as pd

data = {
        'Date': pd.date_range(start='2025/10/01',periods=10,freq='D'),
        'Product':['A','B','C','A','B','C','A','C','B','A'],
        'Units_Sold':[10,15,8,12,18,10,14,9,20,16],
        'Units_Price':[50,40,60,50,40,60,50,60,40,50]
}
df = pd.DataFrame(data)
print(df)
print("üòä_üòä_üòä_üòä_üòä_üòä_üòä_üòä_üòä_üòä_üòä_üòä_üòä_üòä_üòä_üòä_üòä_üòä_üòä_üòä_üòä_üòä")
df['Total_Sales']=df['Units_Sold']*df['Units_Price']
print(df)
print("--------------------------")
Total_revenue = df['Total_Sales'].sum()
avg_Units = df.groupby('Product')['Units_Sold'].mean()
Top_Product = df.groupby('Product')['Total_Sales'].sum().idxmax()



print(f"Total_revenue:{Total_revenue}")
print("--------------------------")
print("\n average units sold per product:\n",avg_Units)
print("--------------------------")
print(f"\n Top Selling product:{Top_Product}")
```

```python
#Example for K Nearest Neighbour(KNN)
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

X = np.array([
    [150,7],     #Apple
    [170,8],     #Apple
    [140,6],
    [130,6],
    [180,9],
    [200,10],
    [210,11],
    [190,9]
])

y = np. array([0,0,0,0,1,1,1,1])

model = KNeighborsClassifier(n_neighbors=3)
model.fit(X,y)

new_fruit = np.array([[160,7]])
prediction = model.predict(new_fruit)

print("Prediction:","Apple" if prediction[0] == 0 else "Orange")

x_min, x_max = X[:, 0].min() - 20, X[:, 0].max() + 20
y_min, y_max = X[:, 1].min() - 2, X[:, 1].max() + 2
xx, yy = np.meshgrid(np.arange(x_min, x_max, 1), np.arange(y_min, y_max, 0.2))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.2)

for i, label in enumerate(y):
  if label == 0:
    plt.scatter(X[i, 0], X[i, 1], marker='o', s=100, label="Apple" if i == 0 else "")
  else:
    plt.scatter(X[i, 0], X[i, 1], marker='s',s=100,label="Orange" if i== 4 else "")
plt.scatter(new_fruit[0, 0], new_fruit[0, 1], marker='*', s=200, label="New Fruit")
plt.xlabel("Weight")
plt.ylabel("Size")
plt.title("KNN: Fruit Classification (Differnt Style")
plt.legend()
plt.show()


```

```python
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

# Training data (Weight of fruits)
X = [[150],[160],[170],[180],[190],[210]]

#Labels:0 = Apple, 1 = Orange
y = [0, 0, 0, 1, 1, 1]

#Create KNN model (K = 3)
model = KNeighborsClassifier(n_neighbors=3)

#Train the model
model.fit(X,y)

#New fruit Weight
new_fruit = [[200]]

#Predict
prediction = model.predict(new_fruit)

#Output result
if prediction[0] == 0:
    print("Prediction: Apple")
else:
    print("Prediction: Orange")

#........................Graph....................


#PLot training data
for i in range(len(X)):
  if y[i] == 0:
    plt.scatter(X[i], 0, marker='o', s=100, label="Apple" if i == 0 else "")
  else:
    plt.scatter(X[i], 0, marker='s', s=100, label="Orange" if i == 3 else "")

#plot new fruit
plt.scatter(new_fruit[0], 0, marker='*', s=200, label="New Fruit")
plt.yticks([]) #Hide Y-axis
plt.xlabel("Weight")
plt.title("Simple KNN Fruit Classification")
plt.legend()
plt.show()


```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X = np.array([[1],[2],[3],[4],[5],[6],[7],[8]])
y = np.array([0,0,0,1,1,1,1,1])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Actual:", y_test)
print("Predicted:", y_pred)
print("Accuracy:", accuracy)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

X = np.array([
    [22,2000],
    [25,2200],
    [23,2100],
    [45,8000],
    [50,8500],
    [48,8200]])

model = KMeans(n_clusters=2)
model.fit(X)
labels = model.labels_
centers = model.cluster_centers_
for i in range(len(X)):
  if labels[i] == 0:
    plt.scatter(X[i][0], X[i][1], color='black')
  else:
    plt.scatter(X[i][0], X[i][1], color='blue')

plt.scatter(centers[:, 0], centers[:, 1],color='green', marker='X', s=200, label='Centers')

plt.xlabel("Age")
plt.ylabel("Monthly Spending")
plt.title("Customer Clustering using K-Means")
plt.legend()
plt.show()

```

```python
#Graph :- 01
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

X = np.array([
    [40,45],
    [42,50],
    [38,43],
    [85,88],
    [87,85],
    [60,65],
    [62,63]])

inertia = []

for k in range(1,6):
  kmeans = KMeans(n_clusters=k, random_state=0)
  kmeans.fit(X)
  inertia.append(kmeans.inertia_)

plt.figure()
plt.plot(range(1,6), inertia, marker='o')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.title("Elbow Method ")
plt.show()



print("----------------------------------------------------------------------------")

#Graph :- 02
model = KMeans(n_clusters=2, random_state=0)
model.fit(X)

labels = model.labels_
centers = model.cluster_centers_

colors = ['blue', 'red', 'green']

plt.figure()
for i in range(len(X)):
  plt.scatter(X[i][0], X[i][1], color=colors[labels[i]])

plt.scatter(centers[:, 0], centers[:, 1], color='black', marker='X', s=200)

plt.xlabel("Maths MArks")
plt.ylabel("Science Marks")
plt.title("Student Clustering")
plt.show()

print("--------------------------------------------------------------------------------------------------")
unique, counts = np.unique(labels, return_counts=True)

plt.figure()
plt.bar(unique, counts,)
plt.xlabel("Cluster Number")
plt.ylabel("Number of Students")
plt.title("Cluster Size Comparison")
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D

X = np.array([
    [40,45,42],
    [42,50,48],
    [38,43,40],
    [85,88,90],
    [90,92,95],
    [87,85,89],
    [60,65,63],
    [62,63,60]])
model = KMeans(n_clusters=3, random_state=42)
model.fit(X)
labels = model.labels_
centers = model.cluster_centers_

plt.figure()
sns.heatmap(X, annot=True, cmap="coolwarm", xticklabels=["Math","Sceince","English"])
plt.title("Student Marks Heatmap")
plt.show()


print("----------------------------------------------------------------------------------------------------")
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for i in range(len(X)):
  ax.scatter(X[i][0], X[i][1], X[i][2], color=colors[labels[i]])

ax.scatter(centers[:, 0], centers[:, 1], centers[:, 2], color='black', marker='X', s=200)

ax.set_xlabel("Math")
ax.set_ylabel("Science")
ax.set_zlabel("English")
ax.set_title(" 3D Student Clustering")
plt.show()






```
